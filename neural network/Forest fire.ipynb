## Import necessary libaries

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
forest_fire = pd.read_csv('forestfires.csv')
forest_fire
```

## Data understanding

```python
forest_fire.shape
```

```python
forest_fire.isna().sum()
```

```python
forest_fire.dtypes
```

### Delete month and day feature

```python
del forest_fire['month']
del forest_fire['day']
```

```python
forest_fire
```

## Label encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
forest_fire['size_category'] = le.fit_transform(forest_fire['size_category'])
```

```python
X = forest_fire.iloc[:,:-1]
y = forest_fire.iloc[:,-1:]
```

```python
y.value_counts()
```

## Standard scalar

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
scaled_X = std.fit_transform(X)
scaled_X = pd.DataFrame(scaled_X,columns=X.columns)
scaled_X
```

## Data split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(scaled_X,y,test_size=0.20,random_state=12,stratify=y)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## Model building

### build architecture

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten,Dense,Dropout
```

```python
model = Sequential()
model.add(Dense(units = 50, activation = 'relu',input_dim = 28)) #I/P
model.add(Dropout(0.5))
model.add(Dense(units = 20, activation = 'relu')) #H1
model.add(Dropout(0.5))
model.add(Dense(units = 1, activation='sigmoid')) #O/P
model.summary()
```

## compile architecture

```python
model.compile(optimizer='Adam',loss ='binary_crossentropy' ,metrics = ['accuracy'])
```

## Train model

```python
model.fit(X_train,y_train,epochs=99)
```

## Model testing

```python
prediction = model.predict(X_test)
```

```python
model.evaluate(X_test,y_test)
```

```python
prediction = np.round(prediction)
```

```python
df = y_test.copy()
df['y_pred'] = prediction
```
